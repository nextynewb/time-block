```python
import pymongo
from pymongo import MongoClient
from datetime import datetime
import time

client = MongoClient('mongodb://localhost:27017/')
db = client['time_block']
col_tasks = db['tasks']

# Function to convert time string to minutes


def time_to_minutes(time_str):
    hours, minutes = map(int, time_str.split(':'))
    return hours * 60 + minutes




# Calculate time spent per category


def calculate_time_per_category():
    categories = {}
    tasks = col_tasks.find({"completed": False})

    for task in tasks:
        category = task['category']
        start_minutes = time_to_minutes(task['start_time'])
        end_minutes = time_to_minutes(task['end_time'])

        # Calculate duration in minutes
        duration = end_minutes - start_minutes
        if duration < 0:  # Handle cases where end time is on the next day
            duration += 24 * 60

        # Add to category total
        if category in categories:
            categories[category] += duration
        else:
            categories[category] = duration
    
    categories = dict(sorted(categories.items(), key=lambda item: item[1], reverse=True))

    for category in categories:
        # Convert minutes to hours and minutes
        hours = categories[category] // 60
        minutes = categories[category] % 60
        categories[category] = f"{hours}h {minutes}m"

    return categories



calculate_time_per_category()
```
